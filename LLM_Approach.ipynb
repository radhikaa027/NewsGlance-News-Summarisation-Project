## Finetuning T5 Model

```python
from datasets import load_dataset
dataset = load_dataset("cnn_dailymail",'3.0.0')
```

```python
model_checkpoint = "t5-small"
max_input_length = 900
max_target_length = 120

tokenizer = AutoTokenizer.from_pretrained(model_checkpoint, use_fast=False, model_max_length = max_input_length)
```

### Preprocess and Tokenize Data


```python
def preprocess_function(examples):
    '''Preprocesses the data as required by huggingface transformer'''
    model_inputs = tokenizer(
        examples["article"], max_length=max_input_length, truncation=True
    )
    with tokenizer.as_target_tokenizer():
        labels = tokenizer(
            examples["highlights"], max_length=max_target_length, truncation=True
        )

    model_inputs["labels"] = labels["input_ids"]
    return model_inputs
```

Sharding the dataset to 1/50th portion to make it smaller, so the training time will reduce.

```python
# Shuffle to get a random sample
dataset.shuffle()
```

```python
dataset_sample = dataset
dataset_sample['train'] = dataset['train'].shard(num_shards=50, index=0)
dataset_sample['validation'] = dataset['validation'].shard(num_shards=40, index=0)
dataset_sample['test'] = dataset['test'].shard(num_shards=40, index=0)
dataset_sample
```

### Tokenize data

We will use a pretrained tokenizer to tokenize the articles and summaries. Each huggingface model has it's own pre trained tokenizer. This tokenizer tokenizes our data in the format required by the model with the pre trained corpus.

The tokenized dataset includes input_ids and attention_mask.

```python
tokenized_datasets = dataset_sample.map(preprocess_function, batched=True)
```

### Loading Pretrained Model for finetuning



```python
model = AutoModelForSeq2SeqLM.from_pretrained(model_checkpoint)
```

```python
batch_size = 8
num_train_epochs = 5
# Show the training loss with every epoch
logging_steps = len(tokenized_datasets["train"]) // batch_size
model_name = model_checkpoint.split("/")[-1]

args = Seq2SeqTrainingArguments(
    output_dir=f"{model_name}-finetuned-cnn-news",
    evaluation_strategy="epoch",
    learning_rate=5.6e-4,
    per_device_train_batch_size=batch_size,
    per_device_eval_batch_size=batch_size,
    weight_decay=0.01,
    save_total_limit=3,
    num_train_epochs=num_train_epochs,
    predict_with_generate=True,
    logging_steps=logging_steps,
    push_to_hub=True,
)
```

```python
import numpy as np
from nltk.tokenize import sent_tokenize

def compute_metrics(eval_pred):
    '''Computes metrics for training'''
    predictions, labels = eval_pred
    decoded_preds = tokenizer.batch_decode(predictions, skip_special_tokens=True)
    labels = np.where(labels != -100, labels, tokenizer.pad_token_id)
    decoded_labels = tokenizer.batch_decode(labels, skip_special_tokens=True)

    decoded_preds = ["\n".join(sent_tokenize(pred.strip())) for pred in decoded_preds]
    decoded_labels = ["\n".join(sent_tokenize(label.strip())) for label in decoded_labels]

    # Compute ROUGE scores
    result = rouge_score.compute(predictions=decoded_preds, references=decoded_labels, use_stemmer=True)
    print(result)  # Check the structure of result

    # Assuming result now contains scalar values, not dictionaries
    # Adjust based on the structure after inspection
    try:
        result = {key: value * 100 for key, value in result.items()}
    except TypeError:
        # Handle case if result is a single scalar value
        result = {"rouge_score": result * 100}

    return {k: round(v, 4) for k, v in result.items()}
```

### Data Collator

```python
from transformers import DataCollatorForSeq2Seq

# Initialize Data Collator
data_collator = DataCollatorForSeq2Seq(tokenizer, model=model)

# Optionally remove existing columns if they exist (useful for original dataset preprocessing)
if "id" in tokenized_datasets["train"].column_names:
    tokenized_datasets = tokenized_datasets.remove_columns(["id", "article", "highlights"])

# Get features for data collator
features = [tokenized_datasets["train"][i] for i in range(2)]
data_collator(features)
```

```python
trainer = Seq2SeqTrainer(
    model,
    args,
    train_dataset=tokenized_datasets["train"],
    eval_dataset=tokenized_datasets["validation"],
    data_collator=data_collator,
    tokenizer=tokenizer,
    compute_metrics=compute_metrics,
)
```

```python
import evaluate
import nltk
from nltk.translate.meteor_score import meteor_score
import numpy as np
from transformers import AutoModelForSeq2SeqLM, AutoTokenizer

# Load metrics
rouge = evaluate.load("rouge")
meteor = evaluate.load("meteor")
bertscore = evaluate.load("bertscore")

# Load your trained summarizer model
model_name = "t5-small"  # or your custom fine-tuned model path
tokenizer = AutoTokenizer.from_pretrained(model_name)
summarizer = AutoModelForSeq2SeqLM.from_pretrained(model_name)

# Initialize accumulators for metrics
rouge1_scores, rouge2_scores, rougel_scores, meteor_scores, bert_scores = [], [], [], [], []

# Loop through each example in the test set
for i in range(len(tokenized_datasets["test"])):
    input_ids = tokenized_datasets["test"]["input_ids"][i]
    reference = tokenizer.decode(tokenized_datasets["test"]["labels"][i], skip_special_tokens=True).strip()

    # Generate summary
    inputs = tokenizer.decode(input_ids, skip_special_tokens=True)
    inputs_encoded = tokenizer.encode(inputs, return_tensors="pt", truncation=True, max_length=900)
    outputs = summarizer.generate(inputs_encoded, max_length=120, num_beams=4, early_stopping=True)
    generated_summary = tokenizer.decode(outputs[0], skip_special_tokens=True).strip()

    # Compute ROUGE scores
    rouge_result = rouge.compute(predictions=[generated_summary], references=[reference])
    rouge1_scores.append(rouge_result['rouge1'])
    rouge2_scores.append(rouge_result['rouge2'])
    rougel_scores.append(rouge_result['rougeL'])

    # Compute METEOR score
    meteor_score_value = meteor_score([reference.split()], generated_summary.split())
    meteor_scores.append(meteor_score_value)

    # Compute BERTScore
    bertscore_result = bertscore.compute(predictions=[generated_summary], references=[reference], lang="en")
    bert_scores.append(np.mean(bertscore_result['f1']))

# Calculate average scores
average_rouge1 = np.mean(rouge1_scores)
average_rouge2 = np.mean(rouge2_scores)
average_rougel = np.mean(rougel_scores)
average_meteor = np.mean(meteor_scores)
average_bertscore = np.mean(bert_scores)

# Compile scores into a dictionary and round to two decimal places
final_scores = {
    "rouge1": round(average_rouge1 * 100, 2),
    "rouge2": round(average_rouge2 * 100, 2),
    "rougeL": round(average_rougel * 100, 2),
    "meteor": round(average_meteor * 100, 2),
    "bertscore": round(average_bertscore * 100, 2)
}

# Print the final scores
print(final_scores)
```
